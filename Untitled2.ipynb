```python
from sklearn import preprocessing
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score

# Dataset
age_group = ['young', 'middle', 'old', 'young', 'middle', 'young', 'old', 
             'middle', 'young', 'middle', 'old', 'young', 'middle', 'old']
income_level = ['low', 'high', 'medium', 'high', 'low', 'medium', 'low', 
                'medium', 'high', 'low', 'medium', 'high', 'medium', 'low']
buy_product = ['no', 'yes', 'no', 'yes', 'no', 'no', 'yes', 
               'yes', 'no', 'yes', 'no', 'yes', 'yes', 'no']

# Encoding categorical data into numerical form
le = preprocessing.LabelEncoder()

age_encoded = le.fit_transform(age_group)  # young=2, middle=1, old=0
income_encoded = le.fit_transform(income_level)  # low=1, medium=2, high=0
buy_encoded = le.fit_transform(buy_product)  # no=0, yes=1

# Combine features
features = list(zip(age_encoded, income_encoded))

# Split dataset into training and testing sets
features_train, features_test, label_train, label_test = train_test_split(
    features, buy_encoded, test_size=0.2, random_state=42
)

# Train the Naive Bayes Classifier
model = GaussianNB()
model.fit(features_train, label_train)

# Predict for the test dataset
predicted_test = model.predict(features_test)

# Evaluate the model
conf_mat = confusion_matrix(label_test, predicted_test)
accuracy = accuracy_score(label_test, predicted_test)

# Output results
print("Test Predictions:", predicted_test)
print("Confusion Matrix:")
print(conf_mat)
print("Accuracy:", accuracy)

# Predict for a new input (e.g., Young, Medium)
new_input = [[le.transform(['young'])[0], le.transform(['medium'])[0]]]  # 'young'=2, 'medium'=2
predicted_new = model.predict(new_input)
print("Prediction for [Young, Medium]:", "Yes" if predicted_new[0] == 1 else "No")
```
